```python
text=['call you tonight','call me a cab','please call me... PLEASE']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vect=CountVectorizer()
```

```python
import pandas  as pd
```

```python
tf=pd.DataFrame(vect.fit_transform(text).toarray(),columns=vect.get_feature_names_out())
#hangi cümlede kaç kere o kelime geçiyor
#vect.fit_transform(text) → text listesindeki metinlere TF-IDF uygular.
#.toarray() → Sonucu NumPy dizisine çevirir.
#pd.DataFrame(..., columns=...) → Sonucu pandas DataFrame’e çevirir, sütun adları olarak kelimeleri kullanır.
#Sonuç: Her satır bir metni, her sütun bir kelimeyi temsil eder; değerler TF-IDF skorudur.
```

```python
tf
```

```python
df=vect.fit_transform(text).toarray().sum(axis=0)
#vect.fit_transform(text) → text verisine TF-IDF uygular.
#.toarray() → TF-IDF matrisini NumPy dizisine çevirir.
#.sum(axis=0) → Her kelimenin tüm metinlerdeki TF-IDF skorlarını toplar.
#Sonuç: Her kelimenin toplam TF-IDF değeri (önemi) hesaplanır.
```

```python
tf/df
```

```python
df=pd.read_csv('yelp.csv')
# 1-Datayı oku
```

```python
df.head()
```

```python
df['text']=df['text'].str.lower()
# 2-Kelimeleri küçük harfe çevir
```

```python
df.head()
```

```python
df['text']=df['text'].str.replace('[^\w\s]','',regex=True)
# 3-Noktalama işaretlerini kaldır
```

```python
df['text']=df['text'].str.replace('\d+','',regex=True)
# 4-Rakamları kaldır
```

```python
df['text']=df['text'].str.replace('\n','',regex=True)
# 5-Satır sonlarını kaldır
```

```python
df['text']=df['text'].str.replace('\r','',regex=True)
# 9-Vektorize et vektöre çevir
```

```python
df['text'][0] 
#dediklerimiz yapılmış
```

```python
#Sentiment Analysis-Bir metnin olumlu, olumsuz veya nötr bir duygu taşıyıp taşımadığını otomatik olarak belirler.
```

```python
ybw=df[(df.stars==1)|(df.stars==5)] 
#1 ve 5 yıldız olanları ayırdık 
```

```python
ybw.head()
```

```python
ybw.reset_index(drop=True,inplace=True)
#bozulan index sıralamsını düzelttik 
```

```python
ybw.head()
```

```python
x=ybw[['text']]
y=ybw[['stars']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
x.head()
```

```python
vect = CountVectorizer(stop_words='english', ngram_range=(1,2)) 
# İngilizce stopword'leri kaldır, 1'li ve 2'li kelime gruplarını say
# 6-Gereksiz kelimeleri kaldır
#9-Vektörize et vektöre çevir
```

```python
xyeni=vect.fit_transform(x['text'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(xyeni, y, random_state=42, test_size=0.20)  
# Veriyi %80 eğitim, %20 test olarak ayır
```

```python
nb = MultinomialNB()  # Naive Bayes modelini oluştur
```

```python
model = nb.fit(x_train, y_train)  # Modeli eğitim verisiyle eğit
```

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```

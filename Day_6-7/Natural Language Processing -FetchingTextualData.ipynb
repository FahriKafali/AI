### Microsoft word dosyası okuma

```python
import docx2txt as wd #word dökümanını okumak için kullanılan sınıf
```

```python
kosey=wd.process('boszaman.docx')
```

```python
for p in kosey.split('\n\n'):
    print(p)
    #paragraf paragraf ayırmış gibi oldu tekte tüm yazıyı tek solukta yazmak yerine
```

```python
dir(kosey) #split komutu gibi neleri kullanavileğine bakmak için kullanılır
```

```python
kosey2=docx.Document('boszaman.docx')
```

### İnternetten Veri alma 

```python
url='https://en.wikipedia.org/wiki/Natural_language_processing'
```

```python
import requests
```

```python
resp=requests.get(url)
```

```python
resp.headers #html de verilen bilgileri getirir
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(resp.text,'lxml')
```

```python
for p in soup.find_all('p'):
    print(p.text) 
#yazılan her şeyi almaya yarar
```

```python
for a in soup.find_all('a'):
    print(a)
    #yazılan tüm linkleri almaya yarar
```

```python
for a in soup.find_all('a'):
    print(a.text)
    #yazılan tüm linkleri yazılarını almaya yarar
```

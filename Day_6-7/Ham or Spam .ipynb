```python
#imbalanced dataset
```

```python
df['class'].value_counts()
```

```python
df['text'] = df['text'].str.lower()  # tüm metni küçük harfe çevir
df['text'] = df['text'].str.replace('[^\w\s]', '', regex=True)  # noktalama işaretlerini kaldır
df['text'] = df['text'].str.replace('\d+', '', regex=True)  # sayıları kaldır
df['text'] = df['text'].str.replace('\n', '', regex=True)  # satır sonlarını sil
df['text'] = df['text'].str.replace('\r', '', regex=True)  # taşıma karakterlerini sil
```

```python
x = df['text']  
y = df['class']
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

pipeline bu yukarıdaki bir fonksiyon haline çevirip direkt kullanılabilir hale getirebiliriz buna da pipline denir boru hattı mimarisi diye geçer 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([('tfidf',TfidfVectorizer()),('classifier',RandomForestClassifier())])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.20, random_state=42) 
# veriyi %80 eğitim, %20 test olacak şekilde ayır
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.svm import SVC
```

```python
svm=Pipeline([('tfidf',TfidfVectorizer()),('classifier',SVC())])
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred=svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
# karışıklık matrisini ısı haritası şeklinde göster, hücrelere sayı yaz
```

```python
print(classification_report(y_test,y_pred))
```

```python
from wordcloud import WordCloud        # kelime bulutu (word cloud) görselleştirmesi için
from wordcloud import STOPWORDS        # yaygın durdurma kelimeleri (stopwords) listesi için
from PIL import Image                  # görsel işleme işlemleri için (resim yükleme vb.)
import numpy as np                     # sayısal işlemler ve matris kullanımı için
import matplotlib.pyplot as plt        # grafik ve görselleştirme işlemleri için
```

```python
def wc(data,color):
    plt.figure(figsize=(10,10))
    #mask=np.array(Image.open('human.jpg'))
    wc=WordCloud(background_color=color,stopwords=STOPWORDS)#,mask=mask sona ekle maske kullanacaksan
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis('off')
```

```python
wc(ham['text'],'white')
```

```python
wc(spam['text'],'black')
```

```python
spam = spam.sample(ham.shape[0], replace=True)  # spam verisinden ham kadar rastgele örnek al (aynı veriyi tekrar seçebilir)

#imbalanced data set olmaktan çıkarttık yani
```

```python
ham.shape,spam.shape
```

```python
data=pd.concat([ham,spam],ignore_index=True)
```

```python
data.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(data['text'], data['class'], test_size=.20, random_state=42)  
# veriyi %80 eğitim, %20 test olarak ayır; giriş: metin, çıktı: sınıf (spam/ham)
```

```python
clf=Pipeline([('tfidf',TfidfVectorizer()),('classifier',RandomForestClassifier())])
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
nlp=spacy.load(name='en_core_web_md')
```

```python
s1='''If the time has come to set sail from life,
A ship departs this harbor, bound for the unknown.

As if it carries no passengers, it silently sets out;
No hands wave, no kerchiefs flutter at its departure.'''
```

```python
print(s1.split()) #tokenize ettik
```

```python
doc1=nlp(s1)
```

```python
#for token in doc1:
 #   print(token)
[x for x in doc1] #üsttekinin aynısını yapıyor
```

```python
s1=[x for x in doc1]
```

```python
[x.lemma_ for x in s1] #ekleri kaldırıp kökleri bulduk 
```

```python
doc=nlp('Apple is looking at buying U.K startuo for $1 billion')
```

```python
for token in doc:
    print(token.text,token.lemma_,token.pos_,token.tag_)
    #özne nesne felan diye ayırdık
```

```python
from spacy import displacy
```

```python
displacy.render(doc,style='dep')
#anlamlandırmak için cümle içinde sıralamalar atlamalar felan 
```
